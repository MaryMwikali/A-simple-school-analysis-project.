```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census_population=pd.read_csv('census_population.csv')
census_population
```

# DATA PREPROCESSING

```python
census_clean=census_population.dropna()
census_clean
```

```python
cleaned_census=census_clean.drop_duplicates()
cleaned_census
```

```python
cleaned_census.columns
```

```python
cleaned_census.dtypes
```

```python
cleaned_census['Category'].count()
```

```python
Category=cleaned_census.pivot_table(cleaned_census,index=['Category'],aggfunc='size')
Category
```

```python
Inhabited_villages=cleaned_census.pivot_table(cleaned_census, index=['Inhabited Villages'],aggfunc='size')
Inhabited_villages
```

```python
Uninhabited_villages=cleaned_census.pivot_table(cleaned_census, index=['Uninhabited Villages'], aggfunc='size')
Uninhabited_villages
```

```python
#positive correlation
Corr=cleaned_census['Inhabited Villages'].corr(cleaned_census['Uninhabited Villages'])
Corr
```

```python
JAMMU=cleaned_census[cleaned_census['Name']=='JAMMU & KASHMIR']
JAMMU
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(JAMMU['Population (Male)'],JAMMU['Population (Female)']);
ax.set(title='JAMMU & KASHMIR POPULATION(MALE VS FEMALE)',
      xlabel='Population(Male)',
      ylabel='Population(Female)');
#correlation between male and female in JAMMU & KASHMIR
```

```python
fig,ax=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax[0].bar(JAMMU['Residence'],JAMMU['Population (Male)']);
ax[1].bar(JAMMU['Residence'],JAMMU['Population (Female)']);
ax[0].set(title='MALE DISTRIBUTION',
         xlabel='Residence',
         ylabel='Population(Male)');
ax[1].set(title='FEMALE DISTRIBUTION',
        xlabel='Residence',
        ylabel='Population(Female)');
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(JAMMU['Residence'],JAMMU['Population Density'])
ax.set(title='RESIDENCE VS POPULATION DENSITY',
      xlabel='Residence',
      ylabel='Population Density');
```

```python
fig,ax=plt.subplots()
ax.bar(cleaned_census['Category'],cleaned_census['Population (Person)'])
ax.set(title='POPULATION IN RESIDENCES',
      xlabel='Category',
      ylabel='POPULATION');
```
